```python
def show_moons(X, y):
    df = pd.DataFrame(dict(x=X[:, 0], y=X[:, 1], label=y))
    colors = {0: 'red', 1: 'midnightblue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    plt.title('Dataset of moons')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label='Class: '+str(key), color=colors[key])
    plt.show()


def show_digits(digits):
    fig = plt.figure(figsize=(6, 6)) 
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    plt.title('Dataset of digits')

    for i in range(64):
        ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
        ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
        ax.text(0, 7, str(digits.target[i]))
    plt.show()
```

### + 4. Зробити висновки про якiсть роботи моделей нейронних мереж на дослi джених даних. В задачах класифiкацiї для кожної навчальної вибiрки вибрати найкращу модель за критерiями якостi.

```python
import warnings
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning
from sklearn.model_selection import GridSearchCV
```

```python
def create_matrix(curr_model, x, y_test):
    y_predict = curr_model.predict(x)
    
    cm = mt.confusion_matrix(y_test, y_predict)
    score = curr_model.score(x, y_test)
    plt.figure(figsize=(9, 9))
    sns.heatmap(cm, annot=True, fmt=".3f", square=True)
    plt.ylabel("Actual label")
    plt.xlabel("Predicted label")
    plt.title(f"Accuracy Score:  {score}", size=15)
    plt.show()
    
    print('Precision-score: ', mt.precision_score(y_test, y_predict, average='micro'))
    print('Recall-score: ', mt.recall_score(y_test, y_predict, average='micro'))
    print('F1-score: ', mt.f1_score(y_test, y_predict, average='micro'))
```

```python

def estimate(curr_model, x, y_true, num_classes):
    create_matrix(curr_model, x, y_true)

    if num_classes == 2:
        pr_curve = mt.plot_precision_recall_curve(curr_model, x, y_true)
        pr_curve.ax_.set_title('Precision-Recall curve')
        plt.show()


        roc_curve_moons = mt.plot_roc_curve(curr_model, x, y_true)
        roc_curve_moons.ax_.set_title('ROC curve')
        plt.show()


    if num_classes > 2:
        y_probs = curr_model.predict_proba(x)
        skplt.plot_precision_recall(y_true, y_probs)
        plt.show()

        skplt.plot_roc(y_true, y_probs)
        plt.show()
```

```python
def create_models(X_train,y_train, X_test, y_test):
    
    hidden_layer_sizes = [(100,), (200,), (100, 50), (50, 100), (50, 100, 50), (100, 50, 100), (50, 50, 50, 50)]
    max_iter = [1000+i*100 for i in range(11)]
    scores = []
    
    for hld in hidden_layer_sizes:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
        scores.clear()
        for mi in max_iter:
            mlp = MLPClassifier(hidden_layer_sizes = hld, max_iter = mi, solver = 'sgd').fit(X_train, y_train)
            scores.append(mlp.score(X_test, y_test))
            
        ax1.set_title(f'hidden_layer_sizes: {hld}')
        ax1.plot(max_iter, scores)
        ax1.set_xlabel('Max iterations')
        ax1.set_ylabel("Model's score")
        
        ax2.set_title(f'Loss curve for hidden_layer_sizes: {hld}')
        ax2.plot(mlp.loss_curve_)
        plt.show()
```

```python
create_models(X_moons_train, y_moons_train, X_moons_test, y_moons_test)
```

З графіків видно, що найкращі результати отримуються для моделей з трьома та чотирьма шарами. Однак врахувуючи відносну простоту та малий обсяг даасетів для подальшої роботи розглядатимуться тришарові.

```python
parameters_m = [{"hidden_layer_sizes": [(50, 100, 50), (100, 50, 100)],
               "activation": ['identity', 'logistic', 'tanh', 'relu'],
               "max_iter": [10000],
               "solver": ['lbfgs', 'sgd', 'adam'],
               "alpha": [10**(i) for i in range(-3, 3)],
              }]
```

### +6. Побудувати ансамбль Voting на основi найпростiших моделей (логістичної регресії, нейронних мереж) 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
```

Параметр voting ансамблю Voting визначає спосіб "голосування" класифікаторів за віднесення і-го прикладу до певного класу.
Можливі значення парметру: 'hard', 'soft'. При значенні параметра 'hard' рішення про остаточний передбачуваний клас деякого прикаду приймається на основі найбільшої кількості голосів за певний клас (тобто, якщо один з класифікаторів спрогнозував клас 0, а 2 інші - клас 1, то остаточним спрогнозованим класом буде 1). При значенні параметра 'soft' рішення про остаточний прогнозований клас приймається на основі найбільшої середньої апостеріорної ймовірності (тобто, якщо для класифікаторів 1, 2, 3 спрогнозовані ймовірності дорівнюють відповідно [0.2, 0.8], [0.8, 0.2], [0.3, 0.7], то остаточний спрогнозований клас буде 1)

При встановленні для параметру voting значення 'soft' можливо також встановити параметр weights, який визначає ступінь "довіри" до класифікатора. Якщо weights не встановлений, то за замовчуванням всі класифікатори рівноцінні і працюють так, як показано в попередньому прикладі.

##### Датасет make_moons:

```python
lg1 = LogisticRegression()
lg2 = LogisticRegression()
mlp1 = MLPClassifier(max_iter = 600)
mlp2 = MLPClassifier(max_iter = 600)

estimators_m = [('log_reg1', lg1),
                ('mlp1', mlp1),
                ('mlp2', mlp2)]
estimators_d = [('log_reg1', lg1),
                ('log_reg2', lg2),
                ('mlp2', mlp2)]

weights_m = [None, [0.2, 0.4, 0.4], [0.5, 0.25, 0.25]]
weights_d = [None, [0.2, 0.4, 0.4], [0.5, 0.25, 0.25]]
```

```python
def decision_regions(X, y, model):
    plot_decision_regions(X, y, clf=model)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('decision regions')
    plt.show()
```

```python
ens_moons_hard = VotingClassifier(estimators = estimators_m,
                                  voting='hard').fit(X_moons_train, y_moons_train)
decision_regions(X_moons, y_moons, ens_moons_hard)
create_matrix(ens_moons_hard, X_moons_test, y_moons_test)
```

```python
for wgt in weights_m:
    ens_moons_soft = VotingClassifier(estimators = estimators_m,
                                      voting='soft',
                                      weights = wgt).fit(X_moons_train, y_moons_train)
    print(ens_moons_soft)
    decision_regions(X_moons, y_moons, ens_moons_soft)    
    estimate(ens_moons_soft, X_moons_test, y_moons_test, 2)
```

```python
ens_digits_hard = VotingClassifier(estimators = estimators_d,
                                  voting='hard').fit(X_digits_train, y_digits_train)
create_matrix(ens_digits_hard, X_digits_test, y_digits_test)
```
